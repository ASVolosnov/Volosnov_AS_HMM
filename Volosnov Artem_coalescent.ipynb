```python
import random
import numpy
```

```python
def w(k, mu, v_t):
    s_1 = 1
    s_2 = 1
    for i in range(1,k+1):
        s_1 *= i
        if i < k-1:
            s_2 *= i
    return k*mu + (1/v_t)*(s_1/(2*s_2))

def Generator(k,mu):
    numpy.random.seed(1)
    mutations = []
    coalescent = []
    t = 0
    k_1 = 0
    k_2 = 0
    while k >= 2:
        v_t = numpy.piecewise(t, (t < 100, t >= 100), (1, 2))
        t += numpy.random.exponential(1/w(k,mu,v_t))
        if numpy.random.binomial(1, k*mu/w(k,mu,v_t)):
            k_1 = random.randint(1,k)
            k_2 = random.randint(1,k)
            while k_1 == k_2:
                k_2 = random.randint(1,k)
            coalescent.append([k_1,k_2,t])
            k -= 1
        else:
            mutations.append([t,random.randint(1,k),random.random()])
    return mutations, coalescent
            
        
```

```python
k = 2
mu = 3
mutations, coalescents = Generator(k,mu)
print("mutations: [время произошедшей мутации, место произошедшей мутации(у какого индивида), в каком месте хромосомы\n",mutations)
print("coalescent: [две позиции у которых произошла коалисенция, время происхождения коалисенции]\n", coalescents)
```
